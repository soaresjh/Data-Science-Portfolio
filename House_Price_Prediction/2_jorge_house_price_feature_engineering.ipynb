```python
import numpy as np
import pandas as pd

from house_prices import to_list_obj, to_list_miss

import seaborn as sbn
import matplotlib.pyplot as plt
```

```python
file = 'train.csv'
```

```python
data = pd.read_csv(file)
```

```python
feat_obj, n_obj = to_list_obj(data, do_print = True)
```

```python
pd.set_option('display.max_columns', n_obj)
data[feat_obj].describe()
```

```python
feat_unique = []
feat_many_uniq = []
is_low_uniq = data[feat_obj].describe().loc['unique'] <= 6


for feat in feat_obj:
    if not is_low_uniq[feat]:
        feat_many_uniq.append(feat)
    else:
        feat_unique.append(feat)
        
print('features with less than 5 uniques %.0f' %len(feat_unique))
print('features with more than 5 uniques %.0f' %len(feat_many_uniq))
```

```python
data[feat_many_uniq].describe()
```

```python
list_house_style = ['1S', '1.5', '2S', '2.5', 'S']
list_uniques = ['1Story', '1.5Fin', '1.5Unf', '2Story', 
                '2.5Fin', '2.5Unf', 'SFoyer', 'SLvl']
print(list_house_style)
print(list_uniques)
```

```python
data['HouseStyle']= \
data['HouseStyle'].map({list_uniques[0]:list_house_style[0],
                        list_uniques[2]:list_house_style[1],
                        list_uniques[1]:list_house_style[1],
                        list_uniques[3]:list_house_style[2],
                        list_uniques[4]:list_house_style[3],
                        list_uniques[5]:list_house_style[3],
                        list_uniques[6]:list_house_style[4],
                        list_uniques[7]:list_house_style[4]
                       })
```

```python
data['RoofMatl'].replace({'WdShake':'Wood', 'WdShngl':'Wood'}, inplace = True)
```

```python
data['Exterior1st'].replace({'AbsShng':'Shing',
                             'AsphShn':'Shing',
                             'BrkComm':'Brick',
                             'Brk Cmn':'Brick',
                             'BrkFace':'Brick',
                             'CBlock':'Cement',
                             'CemntBd':'Cement',
                             'Wd Sdng':'Wood',
                             'Wd Shng':'Wood',
                             'WdShing':'Wood',
                             'Plywood':'Wood'
                            }, inplace = True)
```

```python
data['Exterior2nd'].replace({'AbsShng':'Shing',
                             'AsphShn':'Shing',
                             'BrkComm':'Brick',
                             'Brk Cmn':'Brick',
                             'BrkFace':'Brick',
                             'CBlock':'Cement',
                             'CemntBd':'Cement',
                             'Wd Sdng':'Wood',
                             'Wd Shng':'Wood',
                             'WdShing':'Wood',
                             'Plywood':'Wood'
                            }, inplace = True)
```

```python
list_house_style = ['Typ', 'Min', 'Mod', 'Maj', 'Sev','Sal']
list_uniques = ['Typ', 'Min1', 'Min2', 'Mod', 'Maj1', 'Maj2', 'Sev', 'Sal']
print(list_house_style)
print(list_uniques)
```

```python
data['Functional']= \
data['Functional'].map({list_uniques[0]:list_house_style[0],
                        list_uniques[1]:list_house_style[1],
                        list_uniques[2]:list_house_style[1],
                        list_uniques[3]:list_house_style[2],
                        list_uniques[4]:list_house_style[3],
                        list_uniques[5]:list_house_style[3],
                        list_uniques[6]:list_house_style[4],
                        list_uniques[7]:list_house_style[5]
                       })
```

```python
data.to_csv('x_'+file, index = False)
```
